```python
from sklearn.preprocessing import LabelEncoder
def color_to_int(data):
    le = LabelEncoder()
    le.fit(["green","red","blue","white"])
    data["color"]=le.transform(data["color"]) 
    return data

train_data = color_to_int(nba_train)
test_data = color_to_int(df_test)
```

```python
train_label = pd.get_dummies(nba_train['color'])
test_label = pd.get_dummies(df_test['color'])
pd.DataFrame(train_label)
```

```python
pd.DataFrame(test_label)
```

```python
train_image=train_data.drop(['color'], axis=1)
test_image=test_data.drop(['color'], axis=1)
train_image
```

```python
#資料預處理完成
print('train image:',train_image.shape)
print('train label:',train_label.shape)
print('test image:',test_image.shape)
print('test label:',test_label.shape)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
#建構神經網路
n_input=121
n_hidden_1=64
n_hidden_2=64
n_classes=4
x=tf.placeholder(tf.float32,[None,n_input])
y_=tf.placeholder(tf.float32,[None,n_classes])

with tf.name_scope('hidden1'):
    weights=tf.Variable(tf.random_normal([n_input,n_hidden_1]),name='weights')
    biases=tf.Variable(tf.random_normal([n_hidden_1]),name='biases')
    layer_1=tf.matmul(x,weights)+biases
    
with tf.name_scope('hidden2'):
    weights=tf.Variable(tf.random_normal([n_hidden_1,n_hidden_2]),name='weights')
    biases=tf.Variable(tf.random_normal([n_hidden_2]),name='biases')
    layer_2=tf.matmul(layer_1,weights)+biases

with tf.name_scope('output'):
    weights=tf.Variable(tf.random_normal([n_hidden_1,n_classes]),name='weights')
    biases=tf.Variable(tf.random_normal([n_classes]),name='biases')
    logits=tf.matmul(layer_2,weights)+biases
    predicted = tf.nn.softmax(logits)
```

```python
learning_rate=0.001
total_step=300

#損失函數
with tf.name_scope('cross_entropy'):
    cross_entropy=tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits,labels=y_))
    
with tf.name_scope('train'):
    train_step=tf.train.GradientDescentOptimizer(learning_rate).minimize(cross_entropy)
```

```python
#Accuracy
with tf.name_scope('accuracy'):
    y=tf.nn.softmax(logits)
    correct_prediction=tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
    accuracy=tf.reduce_mean(tf.cast(correct_prediction,'float'))
```

```python
#Train and Test
init=tf.global_variables_initializer()
batch_size=10

with tf.Session() as sess:
    sess.run(init)
    
    
    for i in range(total_step):
        sess.run(train_step,
               feed_dict={x:train_image,y_:train_label})
        
       
        
        if i % 10 == 0:
            loss,acc=sess.run([cross_entropy,accuracy],
                          feed_dict={x:train_image,y_:train_label})
            test_loss,test_acc=sess.run([cross_entropy,accuracy],
                                   feed_dict={x:test_image,y_:test_label})
            print('step:{:4d}\t loss: {:.3f}\t acc:{:.3f}\t test_loss:{:.3f}\t test_acc: {:.3f}'.
                 format(i,loss,acc,test_loss,test_acc))
    print('-------------------------------------------------------------------------\n')
    test_predict=sess.run(predicted,feed_dict={x : test_image})   
    print(test_predict)
```

```python
from sklearn.preprocessing import Binarizer
binarizer=Binarizer(0.5)
test_predict_result=binarizer.fit_transform(test_predict)
test_predict_result=test_predict_result.astype(np.int32)
test_predict_result
```
